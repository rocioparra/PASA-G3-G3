```
# Definiciones de funciones genericas

def get_signal(wav_path):
    _, data = read(wav_path)
    print(data)
    signal = np.array(data,dtype=float) 
    norm = np.linalg.norm(signal)
    print(signal)
    print(norm)
    return signal/np.linalg.norm(signal)  # return normalized version
```

```
# Configuraciones generales
fs = 8000.0
aa = get_signal("pasa-g3-8k-aa.wav")
sh = get_signal("pasa-g3-8k-sh.wav")
```
